# EDGE DETECTION

## Goal : 
- You are transforming a color image 🎨 into grayscale ⚫ and then detecting the edges ✂️ (the outlines of shapes and objects).
- This is a core part of computer vision 🧠 — how a computer “sees” structure.

```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread("image.jpg")
```

### Convert the Image to Grayscale 
`gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)`
- 📘 Explanation:
  - OpenCV reads images in BGR format (Blue, Green, Red 🩵💚❤️).
  - But color isn’t needed to find edges — only brightness 🌓 matters.
  - So we convert it to grayscale, meaning each pixel just has one intensity value (0 = black ⚫ → 255 = white ⚪).

- Why do this?
  - Because edge detection algorithms look for changes in brightness, not color differences.Imagine a mountain 🌄 — the edge is where the light suddenly changes from bright sky to dark slope.

---

### Apply Canny Edge Detection : 
`edges = cv2.Canny(gray, 100, 200)`
- What’s happening here:
  - The Canny algorithm is like a super-smart detective 🕵️‍♀️ for edges.
  - It finds places in the image where brightness changes sharply.

  - `gray` : input grayscale image ⚫
  - `100`  : lower threshold 🔽
  - `200`  : upper threshold 🔼

> Only keep edges where the brightness changes between 100 and 200.

```python
grey = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)

edge = cv2.Canny(grey , 100 , 200)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(grey, cmap='gray')
plt.title('⚫ Grayscale')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edge, cmap='gray')
plt.title('🎯 Detected Edges')
plt.axis('off')

plt.tight_layout()
plt.show()
```
