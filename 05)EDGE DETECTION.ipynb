# EDGE DETECTION

## Goal : 
- You are transforming a color image ðŸŽ¨ into grayscale âš« and then detecting the edges âœ‚ï¸ (the outlines of shapes and objects).
- This is a core part of computer vision ðŸ§  â€” how a computer â€œseesâ€ structure.

```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread("image.jpg")
```

### Convert the Image to Grayscale 
`gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)`
- ðŸ“˜ Explanation:
  - OpenCV reads images in BGR format (Blue, Green, Red ðŸ©µðŸ’šâ¤ï¸).
  - But color isnâ€™t needed to find edges â€” only brightness ðŸŒ“ matters.
  - So we convert it to grayscale, meaning each pixel just has one intensity value (0 = black âš« â†’ 255 = white âšª).

- Why do this?
  - Because edge detection algorithms look for changes in brightness, not color differences.Imagine a mountain ðŸŒ„ â€” the edge is where the light suddenly changes from bright sky to dark slope.

---

### Apply Canny Edge Detection : 
`edges = cv2.Canny(gray, 100, 200)`
- Whatâ€™s happening here:
  - The Canny algorithm is like a super-smart detective ðŸ•µï¸â€â™€ï¸ for edges.
  - It finds places in the image where brightness changes sharply.

  - `gray` : input grayscale image âš«
  - `100`  : lower threshold ðŸ”½
  - `200`  : upper threshold ðŸ”¼

> Only keep edges where the brightness changes between 100 and 200.

```python
grey = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)

edge = cv2.Canny(grey , 100 , 200)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(grey, cmap='gray')
plt.title('âš« Grayscale')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edge, cmap='gray')
plt.title('ðŸŽ¯ Detected Edges')
plt.axis('off')

plt.tight_layout()
plt.show()
```
